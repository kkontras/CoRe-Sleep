```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import base64
import re
%matplotlib inline

# Loading chiesi dataset

file_path = "/users/sista/kkontras/Downloads/AISAH - COPD Baseline Questions.csv"

df = pd.read_csv(file_path, low_memory=False)


print('Table has {} columns and {} datapoints'.format(df.shape[1], df.shape[0]))
```

```python
print("QUESTIONS ASKED IN THE QUESTIONARY", end="\n\n" )
from collections import Counter

for i, col in enumerate(df.columns):
    print("\033[1m {}  \033[0m {}".format(i, col), end="\n\n")
    rows_and_counts = Counter(df[col].to_numpy())
    s = ""
    for i in rows_and_counts:
        s+= " {}:{}   ".format(i, rows_and_counts[i])
    print(s)
    plt.figure()
    plt.hist(np.array(df[col].to_numpy(), dtype=str), len(rows_and_counts))
    plt.xticks(rotation=90)
    plt.show()
```
